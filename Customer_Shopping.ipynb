```python
 df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount', 'frequency_of_purchases' : 'frequency_purchases'})
```

```python
df.columns
```

```python
# create a columns age_group
lables = ['Young_Adult','Adult','Middle_Aged','Senior']
df['Age_Group']=pd.qcut(df['age'],q=4,labels=lables)
```

```python
df[['age','Age_Group']].head(10)
```

```python
# create column purchase_frequency_days
frequency_mapping = {
        'Fortnightly' : 14,
        'Weekly' : 7,
        'Monthly' : 30,
'Quarterly' : 90,
'Bi-Weekly' : 14,
'Annually' : 365,
'Every 3 Months' : 90
}


df['Purchase_frequency_days'] = df['frequency_purchases'].map(frequency_mapping)

```
